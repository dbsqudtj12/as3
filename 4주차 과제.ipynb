Boston House Price

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression, Ridge, Lasso
boston = load_boston()
X=boston.data
y=boston.target
x=[[i[12]] for i in X]
reg=LinearRegression()
lasso_reg=Lasso(alpha=10)
reg.fit(x,y)
lasso_reg.fit(X,y)
```

```python
for a, b in zip(boston.feature_names, lasso_reg.coef_):
    print("%s: %f" %(a,b))
```

```python
reg.score(x,y), lasso_reg.score(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

라쏘 회귀를 이용하여 불필요한 값을 0으로 만든 값이 2번째 코드이다. 
13개의 항목중 절대값이 가장 큰 LSTAT(급여가 낮은 직업에 종사하는 인구비율)에 대해 
회귀한것을 그래프로 표현하였고 score메소드로 모델 성능을 확인 한 결과 0.544정도의 결정계수를 얻었다. 또한 라쏘 회귀를 score메소드로 모델 성능을 확인 한 결과 0.523정도의 결정계수를 얻었다.

당뇨병

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression, Ridge, Lasso
dia = load_diabetes()
X1=dia.data
y1=dia.target
x1=[[i[8]] for i in X1 ]
reg=LinearRegression()
lasso_reg=Lasso(alpha=0.02)
reg.fit(x1,y1)
lasso_reg.fit(X1,y1)
```

```python
for a, b in zip(dia.feature_names, lasso_reg.coef_):
    print("%s: %f" %(a,b))
```

```python
reg.score(x1,y1),lasso_reg.score(X1,y1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x1,y1)
plt.plot(x1,reg.predict(x1),color='red')
```

라쏘 회귀를 이용하여 불필요한 값을 0으로 만든 값이 2번째 코드이다. 10개의 항목중 절대값이 두번째로 큰 S6에 대해 회귀한것을 그래프로 표현하였고 score메소드로 모델 성능을 확인 한 결과 0.32정도의 결정계수를 얻었다. 자료값이 퍼저있어서 결정계수가 상대적으로 낮았고 라쏘 회귀를 이용하면 불필요한 값을 0으로 해서 그런지 score메소드로 모델 성능을 확인한 결과 0.51의 결정계수를 얻었고 선형회귀를 이용한 값보다 모델 성능이 좋은것을 확인 할 수 있다.
